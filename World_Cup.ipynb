Find the correlation between the numerical values


```python
corr_df=df_2022_world_cup_squads_new.corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot= True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title(" Correlation heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

This matrix shows how age, caps, goals, and WC goals are correlated to eachother.

The goals, caps, and WC_goals all have a positive correlation.As a result, the majority of the players scored goals during the World Cup.And, also, the number of goals scored by each player is related to the number of caps that players have.

Also, there is a strong positive correlation between age and caps. This means that as a player's age increases, so does his or her number of caps.
